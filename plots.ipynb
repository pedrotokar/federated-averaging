```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
#Definitivamente não é o melhor jeito de se fazer isso, mas foi o mais conveniente no momento.
nomes = {
    "results/mnist_iid_N100_C0.1_E1_B128.txt": "iid, E = 1, C = 0.1",
    "results/mnist_iid_N100_C0.1_E3_B128.txt": "iid, E = 3, C = 0.1",
    "results/mnist_iid_N100_C0.25_E1_B128.txt": "iid, E = 1, C = 0.25",
    "results/mnist_iid_N100_C0.5_E1_B128.txt": "iid, E = 1, C = 0.5",
    "results/mnist_iid_N100_C0.5_E3_B128.txt": "iid, E = 3, C = 0.5",
    "results/mnist_iid_N100_C0.5_E5_B128.txt": "iid, E = 5, C = 0.5",
    "results/mnist_iid_N100_C0.75_E1_B128.txt": "iid, E = 1, C = 0.75",
    "results/mnist_noniid_N100_C0.1_E1_B128.txt": "non-iid, E = 1, C = 0.1",
    "results/mnist_noniid_N100_C0.1_E3_B128.txt": "non-iid, E = 3, C = 0.1",
    "results/mnist_noniid_N100_C0.1_E5_B128.txt": "non-iid, E = 5, C = 0.1",
    "results/mnist_noniid_N100_C0.1_E10_B128.txt": "non-iid, E = 10, C = 0.1",
    "results/mnist_noniid_N100_C0.1_E20_B128.txt": "non-iid, E = 20, C = 0.1",
    "results/mnist_noniid_N100_C0.1_E50_B128.txt": "non-iid, E = 50, C = 0.1",
    "results/mnist_noniid_N100_C0.25_E1_B128.txt": "non-iid, E = 1, C = 0.25",
    "results/mnist_noniid_N100_C0.5_E1_B128.txt": "non-iid, E = 1, C = 0.5",
    "results/mnist_noniid_N100_C0.5_E3_B128.txt": "non-iid, E = 3, C = 0.5",
    "results/mnist_noniid_N100_C0.75_E1_B128.txt": "non-iid, E = 1, C = 0.75",
}
colunas = dict()
```

```python
for arquivo in nomes:
    hist = []
    with open(arquivo, "r") as f:
        for line in f:
            hist.append(float(line.split(",")[0]))
    colunas[nomes[arquivo]] = hist
df_loss = pd.DataFrame(colunas)
df_loss
```

```python
df_loss.plot()
```

```python
for arquivo in nomes:
    hist = []
    with open(arquivo, "r") as f:
        for line in f:
            hist.append(float(line.split(",")[1]))
    colunas[nomes[arquivo]] = hist
df_acc = pd.DataFrame(colunas)
df_acc
```

```python
df_acc.plot()
```

```python
overfit_test = ["non-iid, E = 1, C = 0.1", "non-iid, E = 5, C = 0.1", "non-iid, E = 10, C = 0.1", "non-iid, E = 20, C = 0.1", "non-iid, E = 50, C = 0.1"]
```

```python
df_acc[overfit_test].plot()
```

```python
c_test_iid = ["iid, E = 1, C = 0.1", "iid, E = 1, C = 0.25", "iid, E = 1, C = 0.5", "iid, E = 1, C = 0.75"]
c_test_noniid = ["non-iid, E = 1, C = 0.1", "non-iid, E = 1, C = 0.25", "non-iid, E = 1, C = 0.5", "non-iid, E = 1, C = 0.75"]
```

```python
fig, axs = plt.subplots(1, 2, figsize = (12, 5), dpi = 100, sharey = True, sharex = True)
axs[0].plot(df_acc[c_test_iid])
axs[0].set_ylim(70, 95)
axs[0].grid(True)
axs[0].yaxis.set_major_formatter(mtick.PercentFormatter())
axs[0].set_xlabel("Round")
axs[0].set_ylabel("Accuracy")
axs[0].legend(c_test_iid)

axs[1].plot(df_acc[c_test_noniid])
axs[1].set_ylim(70, 95)
axs[1].grid(True)
axs[1].yaxis.set_major_formatter(mtick.PercentFormatter())
axs[1].set_xlabel("Round")
axs[1].legend(c_test_noniid)
```
